```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
```

```python

sleep_data = pd.read_csv('data.csv')
sleep_data.head()
```

```python
sleep_data.describe()
```

```python

plt.figure(figsize=(10,7))
sns.lineplot(x=sleep_data["Stress Level"],y=sleep_data["Sleep Duration"])
plt.xlabel("Stress Level")
plt.ylabel("Sleep duration")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sleep_data.boxplot("Sleep Duration","Stress Level")
plt.xlabel("stress level")
plt.ylabel("sleep duration")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x=sleep_data["Age"],y=sleep_data["Sleep Duration"])
plt.xlabel("Age")
plt.ylabel("Sleep duration")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x=sleep_data["BMI Category"],y=sleep_data["Sleep Duration"])
plt.xlabel("BMI Category")
plt.ylabel("Sleep duration")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x=sleep_data["Gender"],y=sleep_data["Sleep Duration"])
plt.xlabel("gender")
plt.ylabel("Sleep duration")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x=sleep_data["Stress Level"],y=sleep_data["Quality of Sleep"])
plt.ylabel("Quality of Sleep")
plt.xlabel("Stress level")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sleep_data.boxplot("Quality of Sleep","Stress Level")
plt.xlabel("stress level")
plt.ylabel("sleep quality")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sleep_data.boxplot("Quality of Sleep","Sleep Disorder")
plt.xlabel("Sleep Disorder")
plt.ylabel("sleep quality")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=sleep_data["Sleep Disorder"],y=sleep_data["Quality of Sleep"])
plt.xlabel("sleep disorder")
plt.ylabel("Quality of Sleep")
plt.show()
```
